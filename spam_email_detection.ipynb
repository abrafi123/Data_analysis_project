```python
import pandas as pd
import numpy as np
import seaborn as sns
import pandas as pd

file_path = r"C:\Users\abuba\OneDrive\Documents\spam_email.csv"

df = pd.read_csv(file_path, header=None)
df = df.iloc[:, :2]
df.columns = ["text", "spam"]

df.head()
```

```python
df['spam'].value_counts()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna().copy()
```

```python
df.isnull().sum()
```

**Seperate X and Y**

```python
df['text'] = df['text'].fillna("").astype(str)
x = df['text'].values
```

```python
y = df['spam'].values
```

**Split dataset**

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

**Data Preprocessing**

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()

xtrain = cv.fit_transform(xtrain)
xtest = cv.transform(xtest)
```

```python
xtrain.toarray()
```

**ML Algorithm**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
Mal = MultinomialNB()
```

```python
Mal.fit(xtrain,ytrain)
```

```python
ypred = Mal.predict(xtest)
```

```python
Mal.score(xtest,ytest)
```
