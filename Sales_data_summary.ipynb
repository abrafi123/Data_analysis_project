```python
#𝙁𝙞𝙜𝙪𝙧𝙚 𝙤𝙪𝙩 𝙏𝙤𝙩𝙖𝙡 𝙎𝙖𝙡𝙚𝙨 
𝒕𝒐𝒕𝒂𝒍_𝒔𝒂𝒍𝒆𝒔 = 0
for 𝒓𝒆𝒄𝒐𝒓𝒅 in 𝒔𝒂𝒍𝒆𝒔:
    𝒕𝒐𝒕𝒂𝒍_𝒔𝒂𝒍𝒆𝒔 = 𝒕𝒐𝒕𝒂𝒍_𝒔𝒂𝒍𝒆𝒔+𝒓𝒆𝒄𝒐𝒓𝒅["amount"]
𝒑𝒓𝒊𝒏𝒕("𝑻𝒐𝒕𝒂𝒍 𝑺𝒂𝒍𝒆𝒔:", 𝒕𝒐𝒕𝒂𝒍_𝒔𝒂𝒍𝒆𝒔)
```

```python
#𝙒𝙝𝙞𝙘𝙝 𝙋𝙧𝙤𝙙𝙪𝙘𝙩 𝙜𝙤𝙩 𝙢𝙤𝙨𝙩 𝙨𝙖𝙡𝙚𝙨 𝙩𝙞𝙡𝙡 𝙣𝙤𝙬
𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔 = {}
for 𝒓𝒆𝒄𝒐𝒓𝒅 in 𝒔𝒂𝒍𝒆𝒔:
    𝒑𝒓𝒐𝒅𝒖𝒄𝒕 = 𝒓𝒆𝒄𝒐𝒓𝒅["product"]
    𝒂𝒎𝒐𝒖𝒏𝒕 = 𝒓𝒆𝒄𝒐𝒓𝒅["amount"]
    if 𝒑𝒓𝒐𝒅𝒖𝒄𝒕 in 𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔:
        𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔[𝒑𝒓𝒐𝒅𝒖𝒄𝒕] += 𝒂𝒎𝒐𝒖𝒏𝒕
    else:
        𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔[𝒑𝒓𝒐𝒅𝒖𝒄𝒕] = 𝒂𝒎𝒐𝒖𝒏𝒕
𝒑𝒓𝒊𝒏𝒕("𝑷𝒓𝒐𝒅𝒖𝒄𝒕 𝑾𝒊𝒔𝒆 𝑺𝒂𝒍𝒆𝒔:", 𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔)
𝒑𝒓𝒊𝒏𝒕("𝑴𝒐𝒔𝒕 𝑺𝒐𝒍𝒅 𝑷𝒓𝒐𝒅𝒖𝒄𝒕:", 𝒎𝒂𝒙(𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔, 𝒌𝒆𝒚=𝒑𝒓𝒐𝒅𝒖𝒄𝒕_𝒔𝒂𝒍𝒆𝒔.𝒈𝒆𝒕))
```

```python
#Grapically representation
import matplotlib.pyplot as plt

sales = [
    {'day': 'Monday', 'product': 'Shirt', 'amount': 1200},
    {'day': 'Tuesday', 'product': 'Pant', 'amount': 800},
    {'day': 'Monday', 'product': 'Shoe', 'amount': 500},
    {'day': 'Wednesday', 'product': 'Shirt', 'amount': 700},
    {'day': 'Thursday', 'product': 'Pant', 'amount': 600},
]

product_sales = {}
for record in sales:
    product = record["product"]
    amount = record["amount"]
    if product in product_sales:
        product_sales[product] += amount
    else:
        product_sales[product] = amount

print("Product Wise Sales:", product_sales)
print("Most Sold Product:", max(product_sales, key=product_sales.get))

#bar chart
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(product_sales.keys(), product_sales.values(), color="skyblue", edgecolor="black")
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Product Wise Sales (Bar Chart)")

# Pie Chart
plt.subplot(1,2,2)
plt.pie(product_sales.values(), labels=product_sales.keys(), autopct="%1.1f%%", startangle=90)
plt.title("Product Wise Sales (Pie Chart)")

plt.tight_layout()
plt.show()
```

```python
#Grapically representation
import matplotlib.pyplot as plt

sales = [
    {'day': 'Monday', 'product': 'Shirt', 'amount': 1200},
    {'day': 'Tuesday', 'product': 'Pant', 'amount': 800},
    {'day': 'Monday', 'product': 'Shoe', 'amount': 500},
]

day_sales = {}
for record in sales:
    day = record['day']       
    amount = record['amount']
    
    if day in day_sales:
        day_sales[day] += amount
    else:
        day_sales[day] = amount

print("Day Wise Sales:", day_sales)
print("Highest Sales Day:", max(day_sales, key=day_sales.get))

plt.bar(day_sales.keys(), day_sales.values())
plt.xlabel("Days")
plt.ylabel("Total Sales")
plt.title("Day Wise Sales")
plt.show()
```
